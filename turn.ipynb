```python
import pandas as pd
import re
import string
csv_pdf=pd.read_csv("easa_faqs.csv")
```

```python
csv_pdf.columns
```

```python
csv_pdf = csv_pdf.loc[:, ~csv_pdf.columns.str.contains('^Unnamed')]
```

```python
def clean_text(text):
    text=str(text)
    text=re.sub('\[*?\]', '', text)
    text=re.sub('https?://\S+|www\.\S+', '', text)
    text=re.sub('<.*?>+', '', text)
    text = re.sub('\n', '', text)
    return text
```

```python
csv_pdf.columns
```

```python
csv_pdf['question']=csv_pdf['question'].apply(clean_text)
```

```python
csv_pdf['category']=csv_pdf['category'].apply(clean_text)
```

```python
csv_pdf['answer']=csv_pdf['answer'].apply(clean_text)
```

```python
csv_pdf.to_csv("clean_easa_faqs.csv")
```
